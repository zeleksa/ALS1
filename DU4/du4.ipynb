```python
from r_tree import R_tree, Index_Record, Rectangle
```

```python
# Maximum amount of records stored in one node
M = 3

# Rectangles to be inserted
rectangles = [
    Rectangle(1, 2, 1, 2),
    Rectangle(3, 4, 3, 4),
    Rectangle(2, 5, 1, 3),
    Rectangle(6, 7, 5, 6),
    Rectangle(1, 5, 2, 4),
    Rectangle(10, 12, 6, 8),
    Rectangle(5, 7, 5, 6),
    Rectangle(1, 3, 7, 9),
    Rectangle(20, 23, 15, 19),
    Rectangle(1, 4, 2, 3),
    Rectangle(1, 4, 1, 4),
    Rectangle(1, 3, 5, 7),
    Rectangle(1, 5, 2, 3),
    Rectangle(1, 4, 2, 3)
]

# Rectangle ids
key_ides = [i for i in range(len(rectangles))]
```

```python
# Make R tree
rt = R_tree(M)

# Insert rectangles into R tree
for i in range(len(rectangles)):
    rt.insert(Index_Record(rectangles[i], key_ides[i]))
    rt.print_tree(rt.root, 0)
    print()
    rt.plot_tree()
```

```python
# Test if any index record is missing
miss_test = rt.missing_test(rt.root, key_ides)
if all(miss_test):
    print("OK")
else:
    print("Something is missing")
    print(miss_test)
```
